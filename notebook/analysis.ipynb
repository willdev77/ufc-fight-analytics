```python
df["STR_diff"] = df["STR_1"] - df["STR_2"]
df["TD_diff"] = df["TD_1"] - df["TD_2"]
df["KD_diff"] = df["KD_1"] - df["KD_2"]
```

```python
sns.histplot(df["STR_1"], bins=30)

plt.title("Distribuição de golpes significativos")
plt.show()
```

```python
style_cols = ["Distance_%_1","Clinch_%_1","Ground_%_1"]

df[style_cols].mean()
```

```python
df[style_cols].mean().plot(kind="bar")

plt.title("Distribuição média de estilos de luta")
plt.show()
```

```python
sns.scatterplot(x=df["STR_1"], y=df["KD_1"])

plt.title("Golpes vs Knockdowns")
plt.show()
```

```python
sns.histplot(df["Sig. Str._%_1"])

plt.title("Precisão de golpes significativos")
plt.show()    
```

```python
top_strikers = df.groupby("Fighter_1")["STR_1"].mean().sort_values(ascending=False).head(10)

top_strikers
```

```python
top_strikers.plot(kind="bar")

plt.title("Lutadores com maior média de golpes")
plt.show()
```

```python
avg_strikes = df[["STR_1","STR_2"]].mean()

avg_strikes.plot(kind="bar")

plt.title("Média de golpes significativos por lutador")
plt.ylabel("Quantidade de golpes")
plt.show()
```

```python
sns.histplot(df["Sig. Str. %_1"], bins=30)

plt.title("Distribuição de precisão de golpes")
plt.xlabel("Precisão")
plt.show()
```

```python
style = df[["Distance_%_1","Clinch_%_1","Ground_%_1"]].mean()

style.plot(kind="bar")

plt.title("Distribuição média dos estilos de luta")
plt.ylabel("Proporção da luta")
plt.show()
```

```python
sns.scatterplot(x=df["STR_1"], y=df["KD_1"])

plt.title("Relação entre golpes e knockdowns")
plt.xlabel("Golpes significativos")
plt.ylabel("Knockdowns")
plt.show()
```

```python
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)

top_strikers = (
    df.groupby("Fighter_1")["STR_1"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

top_strikers.plot(kind="barh")

plt.title("Top 10 lutadores com maior média de golpes")
plt.xlabel("Golpes por luta")
plt.show()
```
